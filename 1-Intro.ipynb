```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_uint
from skimage.io import imshow,imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

Basic Image

```python
array_1 = np.array([
    [255,0],[0,255]
])

imshow(array_1,cmap='gray')
```

Create another one with the matrix with is 3x3 

```python
array_2 = np.array([
    [255,0,255],
    [0,255,0],
    [255,0,255]
])

imshow(array_2,cmap='gray')
```

Create with the arrange function, with display a shade of gray in step of 17

```python
array_spectrun = np.array([
    np.arange(0,255,17),
    np.arange(255,0,-17),
    np.arange(0,255,17),
    np.arange(255,0,-17)
])

fig,ax = plt.subplots(1,2,figsize=(12,4))

ax[0].imshow(array_spectrun,cmap='gray')
ax[0].set_title('Arange Image')

# Can create another one with the transpose
ax[1].imshow(array_spectrun.T,cmap='gray')
ax[1].set_title("Transpose Image")
```

Create with RGB color

```python
array_colors = np.array([
    [
        [255,0,0],[0,255,0],[0,0,255]
    ]
])

imshow(array_colors)
```

Doing with actual image

```python
myimage = imread('example.jpg')
imshow(myimage)
```

Split into 3 sections

```python
fig,ax = plt.subplots(1,3,figsize=(6,4),sharey=True)

ax[0].imshow(myimage[:,0:227])
ax[0].set_title('First Split')

ax[1].imshow(myimage[:,228:456])
ax[1].set_title('Second Split')

ax[2].imshow(myimage[:,457:683])
ax[2].set_title('Third Split')
```

```python

torei = myimage[100:520,290:683]
imshow(torei) 
```

Divide the color mixing

```python
fig,ax = plt.subplots(1,3,figsize=(12,4),sharey=True)

ax[0].imshow(myimage[:,:,0],cmap="Reds")
ax[0].set_title("Red")
ax[1].imshow(myimage[:,:,1],cmap="Greens")
ax[1].set_title('Greens')
ax[2].imshow(myimage[:,:,2],cmap="Blues")
ax[2].set_title('Blue')
```

Convert to HSV with Hue, Saturation, and Value

```python
torei_hsv = rgb2hsv(myimage)
fig,ax = plt.subplots(1,3,figsize=(12,4),sharey=True)

ax[0].imshow(torei_hsv[:,:,0],cmap='hsv')
ax[0].set_title('Hue')

ax[1].imshow(torei_hsv[:,:,1],cmap='hsv')
ax[1].set_title('Saturation')

ax[2].imshow(torei_hsv[:,:,2],cmap='hsv')
ax[2].set_title('Value')
```

Convert to a grayscale image

```python
torei_gray = rgb2gray(myimage)

fig,ax = plt.subplots(1,5,figsize=(17,6),sharey=True)
ax[0].imshow(torei_gray,cmap='gray')
ax[0].set_title('Grayscale Image')

ax[1].imshow(img_as_uint(torei_gray > 0.25),cmap='gray')
ax[1].set_title('Greater Than 0.25')

ax[2].imshow(img_as_uint(torei_gray >0.50),cmap='gray')
ax[2].set_title('Greater Than 0.5')

ax[3].imshow(img_as_uint(torei_gray>0.75),cmap='gray')
ax[3].set_title('Grater Than 0.75')

ax[4].imshow(img_as_uint(torei_gray > np.mean(torei_gray)),cmap='gray')
ax[4].set_title("Greater Than Mean")
```
