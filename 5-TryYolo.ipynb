## Trying to use YoLo 
With the tutorial and pre-trained model

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg


# Use the pretrained YOLOv8 n model which is smallest
model = YOLO('yolov8n.pt')
```

Testing with image object detection

```python
source_image_path = 'data/sample-image/market-1.jpeg'
img = mpimg.imread(source_image_path)

fig,ax = plt.subplots(1,2,figsize=(12,6))
ax[0].imshow(img,cmap=plt.cm.binary)

results = model.predict(source_image_path)
annotated_result = results[0].plot()
ax[1].imshow(annotated_result)
```
